# plot ECDF and survival curves(norm=True) based on exit times

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

EXIT_THRESH = 0.2

fig = plt.figure(figsize=(10,5))
grid = fig.add_gridspec(1, 6)

ax1 = fig.add_subplot(grid[:, 0:2])
ax2 = fig.add_subplot(grid[:, 2:6])


def get_exit_times(mu=0, sigma=0.06, xi=0.1, tmax=20, N=10000):
    dt = 0.001
    exit_times = np.zeros(N)
    mu_ = np.random.normal(mu, scale=sigma, size=N)
    x = np.repeat(-2.0, N)
    
    for t in np.arange(0, tmax, dt):
        x += (np.power(x, 2) + mu_) * dt + xi * np.sqrt(dt) * np.random.normal(size=N)
        exit_times[(exit_times == 0) & (x > EXIT_THRESH)] = t
        x[x > EXIT_THRESH] = 1
        
        if (exit_times > 0).all():
            break
            
    exit_times[exit_times == 0] = tmax
    return exit_times

def generate_antibody_titers(tmax=20, SAMPLE_RATE=0.1, N=10000):
    exit_times = get_exit_times(tmax=tmax, N=N)
    surv = []
    t_out = []
    surv_fact = 1
    
    # Generate time points for which survival fraction will be recorded
    time_points = np.arange(0, tmax, SAMPLE_RATE)
    
    for t in time_points:
        # Compute the proportion of particles that haven't exited by this time point
        surv_fact = np.mean(exit_times > t)
        t_out.append(t)
        surv.append(surv_fact)
               
    return np.array(t_out), np.array(surv)
    

def plotExitTimesSampled(ax, t_out, surv, tmin=0, tmax=20, c='r', norm=True, label=None, ls='solid'):
    x = np.linspace(tmin, tmax, 1000)
    y = interp1d(t_out, 1200*surv)(x) # set inital number of plasma cells to be 1200
    if norm:
        log_y = np.log(y)
        y = np.exp((log_y - log_y.min()) / (log_y.max() - log_y.min())) / np.exp(1)
    ax.plot(x, y, c=c, label=label)

#plot ECDF
exit_times = get_exit_times()
sorted_exit_times = np.sort(exit_times)
cdf = np.arange(1, len(sorted_exit_times) + 1) / len(sorted_exit_times)

ax1.step(sorted_exit_times, cdf, where='post')
ax1.set_xlabel("Exit Time")
ax1.set_ylabel("CDF")
ax1.set_title("Empirical CDF of Exit Times")

#plot antibody titers vs years after vaccination
num_simulations = 5 # Number of simulations
datasets = []
colors = plt.cm.viridis(np.linspace(0, 1, num_simulations)) # Colors for different simulations

for i in range(num_simulations):
    t_out, surv = generate_antibody_titers()
    datasets.append((t_out, surv))
    
for i, (t_out, surv) in enumerate(datasets):
    plotExitTimesSampled(ax2, t_out, surv, tmin=0, tmax=10, c=colors[i], norm=True, label=f'Simulation {i+1}')
    
ax2.set_xlabel('years after vaccination')
ax2.set_ylabel('antibody titers (log norm.)')
ax2.legend()

fig.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=1)

plt.tight_layout()

plt.show()

```

# plot ECDF and survival curves(norm=True) based on state "x"

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

EXIT_THRESH = 0.2

fig = plt.figure(figsize=(10,5))
grid = fig.add_gridspec(1, 6)

ax1 = fig.add_subplot(grid[:, 0:2])
ax2 = fig.add_subplot(grid[:, 2:6])


def get_exit_times(mu=0, sigma=0.06, xi=0.1, tmax=20, N=10000):
    dt = 0.001
    exit_times = np.zeros(N)
    mu_ = np.random.normal(mu, scale=sigma, size=N)
    x = np.repeat(-2.0, N)
    
    for t in np.arange(0, tmax, dt):
        x += (np.power(x, 2) + mu_) * dt + xi * np.sqrt(dt) * np.random.normal(size=N)
        exit_times[(exit_times == 0) & (x > EXIT_THRESH)] = t
        x[x > EXIT_THRESH] = 1
        
        if (exit_times > 0).all():
            break
            
    exit_times[exit_times == 0] = tmax
    return exit_times, x

# convert to antibody titers   
def get_exit_idx(x):
    return x > EXIT_THRESH

def generate_antibody_titers(tmax=20, SAMPLE_RATE=0.1):
    exit_times, x = get_exit_times()
    surv = []
    t_out = []
    surv_fact = 1
    for t in np.arange(0, tmax, SAMPLE_RATE):
        if t % SAMPLE_RATE < 1e-9:
            idx = get_exit_idx(x)
            surv_fact *= 1 - idx.mean()
            t_out = np.append(t_out, t)
            surv = np.append(surv, surv_fact)               
    return t_out, surv 

def plotExitTimesSampled(ax, t_out, surv, tmin=0, tmax=20, c='r', norm=True, label=None, ls='solid'):
    x = np.linspace(tmin, tmax, 1000)
    y = interp1d(t_out, 1200*surv)(x)
    if norm:
        log_y = np.log(y)
        y = np.exp((log_y - log_y.min()) / (log_y.max() - log_y.min())) / np.exp(1)
    ax.plot(x, y, c=c, label=label)

#plot ECDF
exit_times = get_exit_times()[0]
sorted_exit_times = np.sort(exit_times)
cdf = np.arange(1, len(sorted_exit_times) + 1) / len(sorted_exit_times)

ax1.step(sorted_exit_times, cdf, where='post')
ax1.set_xlabel("Exit Time")
ax1.set_ylabel("CDF")
ax1.set_title("Empirical CDF of Exit Times")

#plot antibody titers vs years after vaccination
num_simulations = 5 # Number of simulations
datasets = []
colors = plt.cm.viridis(np.linspace(0, 1, num_simulations)) # Colors for different simulations

for i in range(num_simulations):
    t_out, surv = generate_antibody_titers()
    datasets.append((t_out, surv))
    
for i, (t_out, surv) in enumerate(datasets):
    plotExitTimesSampled(ax2, t_out, surv, tmin=0, tmax=10, c=colors[i], norm=True, label=f'Simulation {i+1}')
    
ax2.set_xlabel('years after vaccination')
ax2.set_ylabel('antibody titers (log norm.)')
ax2.legend()

fig.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=1)

plt.tight_layout()

plt.show()
```

# Cells(PDF) from Star Methods

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

std_val = 0.06
parameters_human = {"sigma":0.1}

x = np.linspace(-5,5,1000)
pdf = norm.pdf(x, loc=0, scale=std_val / (parameters_human["sigma"] ** (4 / 3)))
cdf = norm.cdf(x, loc=0, scale=std_val / (parameters_human["sigma"] ** (4 / 3)))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

# Plot PDF
ax1.plot(x, pdf, c='k', lw=3)
ax1.spines['left'].set_position('zero')
ax1.spines['right'].set_color('none')
ax1.spines['top'].set_color('none')
ax1.set_yticks([])
ax1.set_xticks([-5, -2.5, 0, 2.5, 5])
ax1.set_xlabel(r"$\mu \xi^{-4/3}$")
ax1.set_ylabel(r"cells (PDF)")

# Plot CDF
ax2.plot(x, cdf, c='k', lw=3)
ax2.spines['left'].set_position('zero')
ax2.spines['right'].set_color('none')
ax2.spines['top'].set_color('none')
ax2.set_yticks([])
ax2.set_xticks([-5, -2.5, 0, 2.5, 5])
ax2.set_xlabel(r"$\mu \xi^{-4/3}$")
ax2.set_ylabel(r"cells (CDF)")

plt.show()
```
